```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\suruj\PYTHON-IMAGE-PROCESSING\demo\pht\suruj.jpg")
img = cv2.resize(img,(450,650))  # width height
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


img = cv2.rectangle(img , (140,120) , (220 ,240) , (0 , 0 , 0) , 0)
#plt.imshow(img)       
b,g,r = cv2.split(img)
img = cv2.merge((r,g,b))


roi = img[120:240,140:220]  # pass the format -- (y1:y2) , (x1,x2)  #calculate the euclidean distance x = (220-140=80) y = (240-120=120)
#now passing data into img
img[120:240,220:300] = roi # right-->
img[120:240,300:380] = roi # right -->
img[120:240,60:140] = roi  #left -->
#img[250:370,170:250] = roi # format img[y1:y2,x1:x2]
# cv2.imshow("origin", img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
cv2.imshow("ravan" , img)
b,g,r = cv2.split(img)
img = cv2.merge((r,g,b))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
