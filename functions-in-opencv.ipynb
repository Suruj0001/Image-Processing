```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread(r"C:\Users\suruj\machine-learning bootcamp\photos\suruj.jpg")
```

```python
plt.imshow(img1)
```

```python
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
#--line
img2 = cv2.line(img2 , (0,0) , (1000 , 1000) , (245 , 245 , 31) , 18) 
```

```python
plt.imshow(img2)
```

```python
# arrowed line -- 

img3 = cv2.arrowedLine(img2 , (0,0) , (1000 , 1000) , (245 , 245 , 31) , 18)
```

```python
plt.imshow(img3)
```

```python
img4 = cv2.rectangle(img1 , (1000,500) , (2000 , 2000) , (245 , 245 , 31) , 80) 
img4 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img4)
```

```python
img6 = cv2.imread(r"C:\Users\suruj\machine-learning bootcamp\photos\suruj.jpg")
img6 = cv2.cvtColor(img6, cv2.COLOR_BGR2RGB)
font = cv2.FONT_HERSHEY_SIMPLEX
img6 = cv2.putText(img6 , 'I AM SURUJ' , (500 , 1000) , font , 12 , (552 , 0 , 2) , 15 , cv2.LINE_AA)   #coor --x , --y (500 , 1000)
```

```python
#ellipse---
img7 = cv2.imread(r"C:\Users\suruj\machine-learning bootcamp\photos\suruj.jpg")
img7 = cv2.cvtColor(img7, cv2.COLOR_BGR2RGB)
#-------------------------START-COOR--LENGTH-HEIGHT---COLOR--THICKNESS --ROTATION-POINT(X,Y)---PARAMETER(ANGLE)-COLOR-CODE---THICKNESS
img7 = cv2.ellipse(img7 ,(500,1500) , (1000,500) , 0,0,180,9,18) 
```

```python
plt.imshow(img7)
```
