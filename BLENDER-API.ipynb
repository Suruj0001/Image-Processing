```python
img1 = cv.imread(r"C:\Users\suruj\PYTHON-IMAGE-PROCESSING\demo\pht\suruj.jpg")
img1 = cv.resize(img1,(500,500))
img2 = cv.imread(r"C:\Users\suruj\PYTHON-IMAGE-PROCESSING\demo\pht\lion.jpg")
img2 = cv.resize(img2,(500,500))
#cv.imshow("First-Image",img1)
#cv.imshow("Second-Image",img2)
```

```python
while True:
    s = cv.getTrackbarPos(switch,"win")
    a = cv.getTrackbarPos("alpha","win")
    n = float(a/100)
    print(n)
   


     
    if s ==0:
        dst = img[:]
    else:
        dst = cv.addWeighted(img1,1-n,img2,n,0)
        cv.putText(dst, str(a), (20,50), cv.FONT_ITALIC,2,(0,125,255),2)
        
    cv.imshow('BLENDER-API',dst)
    k=cv.waitKey(1) 
    if k == ord('q'):
        break
```
